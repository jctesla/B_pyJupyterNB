```python
# como crear un reverso de una LISTA
s = '0123456789abcdef'
d = list(s)
d.reverse()
s = ''.join(c for c in d)
print(s)
```

```python
# Howto Open a FILE I/O
f=open('fichTrabajo.txt', 'r')
while True:
    L = f.read()
    f.seek(5)
    if len(L) < 1:
        f.close()
        break
    print(L)    
```

```python
f = open('fichTrabajo.txt', 'rb+')
f.write(b'ABCDfedcba987643210')
f.seek(5)                                 # Go to the 6th byte in the file, begin idx=0
r = f.read(1)
print(f'Lei del file r={r}')
f.close()
```

```python
f=open('fichTrabajo.txt', 'w')
f.append('Probando, probando\n')             # escribe el contenido de cadena al fichero y devuelve None.
valor = f'La Respuesta es 42'
f.write(valor)
f.close()
```
