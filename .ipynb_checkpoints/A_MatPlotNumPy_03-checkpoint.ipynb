```python
# quiero diagramar el resultado de esta ecuacion
# si el resultado de pY > 10.
pX = [-1,-2,-3,1,2,3,4,5]
pY = []

# Evaluo el resultado de la Ecuacion 
# con los Valores de pX, si la Ecuacion > 10.
for vv in pX:
    rslt = lambda x : x**2 + 6
    xx = rslt(vv)
    #print(xx)
    if (xx > 10):
       pY.append(xx) 
    else:
       pY.append(0) 
    
print(f'El valor de pY = {pY}')
print(f'El valor de pX = {pX}')
from pylab import plot, show
plot(pX, pY)
show()

#pero si quitamos los valores = '0'
ppY=[]
ppX=[]
for idx,vlr in enumerate(pY):
    if vlr > 0:
        ppY.append(vlr)
        ppX.append(pX[idx])
        
print(f'valor de ppY={ppY}')
print(f'valor de ppX={ppX}')
plot(ppX, ppY)
show()
```

 <h1  style="background-color:powderblue;text-align:center;">CIRCULO</h1)

```python
# ecuacionde un circulo (x-a)²+(y-b)² = r²
# conocemos el radio y un eje= x
# despejamos la funcion f(x) = raiz{r² - (x-a)²} + b

from math import sqrt

# contanste q tengo
r=2
a=0
b=0

# genero un  rango aleatorio cualquiera
eje_x = [x/100 for x in range(-200,200)]
#print(eje_x)

# Ecuacion de calculo de y, si tengo r y x
ecu_y = lambda x,r,a,b,s : s * (sqrt((r**2) - (x-a)**2) + b)
eje_pos_y = [ecu_y(n,r,a,b,1) for n in eje_x]
eje_neg_y = [ecu_y(n,r,a,b,-1) for n in eje_x]

from pylab import plot, show
plot(eje_x, eje_pos_y)
plot(eje_x, eje_neg_y)
show()
```

```python
#Tabla de Equivalencia de Radianes, Grados 
import math

INPUTS = [
            (0, 0),
            (math.pi / 6, 30),
            (math.pi / 4, 45),
            (math.pi / 3, 60),
            (math.pi / 2, 90),
            (math.pi, 180),
            (3 * math.pi / 2, 270),
            (2 * math.pi, 360),
         ]

print(f'Radians    Degrees    Expected')
print('{:-^8} {:-^8} {:-^8}'.format('', '', ''))
for rad, expected in INPUTS:
    print('{:8.2f} {:8.2f} {:8.2f}'.format(
        rad,
        math.degrees(rad),
        expected,
    ))
```

```python
# Derivada de Seno
import math
ecuOrig = lambda x : math.sin(math.radians(x))        # Seno convierto x a grados
ecuDeriv= lambda x : math.cos(math.radians(x))        # derivada de Seno 
fncOrigX = [ecuOrig(n) for n in range(-180,180)]
fncDerivX = [ecuDeriv(n) for n in range(-180,180)]
funcY = [n for n in range(-180,180)]
plot(funcY,fncOrigX)
plot(funcY,fncDerivX)
show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec


def label_axes(fig):
    for i, plott in enumerate(fig.axes):      # axes son los plts creados 1,2,3,4
        plott.text(0.5, 0.5, "plot%d" % (i+1), va="center", ha="center")
        plott.tick_params(labelbottom=False, labelleft=False)


fig = plt.figure(1)
fig.suptitle("sizes width & height")

# Crea una Tabla de 2 por 2 grillas --> GridSpec(2,2); El tamaño de las 2 grillas es:
# en ancho -> width =[1, 2], en proporcion la 1ra la mitad de la 2da = [1,2]
# en alto  -> height=[2, 1], en proporcion la 1ra grilla el Doble de la 2da = [2,1]
grd = GridSpec(2, 2, width_ratios=[1, 2], height_ratios=[2, 1])
#grd = GridSpec(1, 2)

x = [0, 1, 2, 3]
y = [3, 8, 1, 10]

# 2x2 grillas son 4 sub tablas=grillas p plotear.
fig.add_subplot(grd[0])
plt.plot(x,y)

fig.add_subplot(grd[1])
plt.plot(x,y)

fig.add_subplot(grd[2])
# plt.plot(x,y)

fig.add_subplot(grd[3])
plt.plot(x,y)

label_axes(fig)

#---SHOW----------
plt.show()
```
