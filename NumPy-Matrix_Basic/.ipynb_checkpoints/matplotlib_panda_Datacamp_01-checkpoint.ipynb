```python
#Operaciones con la Lista q se traiga, una SUMA
df = pd.read_excel('D:/phytonSpace/DeepLearning/DataCamp/Hillary MML/tblPanda_01.xlsx', 'ejemplo_01')
df_rqst = df['cod Ing'] 
print('suma valores df = ',df_rqst.sum())
print('valos Maximo df =',df_rqst.max())
print('valos Minimo df =',df_rqst.min())
print('valos id Maximo df =',df_rqst.idxmax())
print('valos id Minimo df =',df_rqst.idxmin())
```

```python
# Rqst CONDICIONALES USANDO OPERADORES LOGICOS;
df_rqst = pd.read_excel('D:/phytonSpace/DeepLearning/DataCamp/Hillary MML/tblPanda_01.xlsx', 'ejemplo_01')

reporte = df_rqst[df_rqst.price > 15]
print('price > 15 \n', reporte,'\n')

reporte = df_rqst[df_rqst.frequency > 3] 
print('frequency > 3 \n', reporte,'\n')

reporte = df_rqst[df_rqst.frequency == 1.37]
print('frequency == 1.37 \n', reporte,'\n')

reporte = df_rqst[(df_rqst.frequency != 7.38) & (df_rqst.frequency != 2.46)]
print('frequency != 7.38 and frequency != 2.46  \n', reporte,'\n')

reporte = df_rqst[(df_rqst.frequency == 7.38) | (df_rqst.frequency == 2.46)]
print('frequency = 7.38 or frequency = 2.46  \n', reporte,'\n')
```

```python
# CONDICIONAL sin usar los operadores de condicional rqst > 105
df = pd.read_excel('D:/phytonSpace/DeepLearning/DataCamp/Hillary MML/tblPanda_01.xlsx', 'ejemplo_01')
rqst = df['cod Ing']
listaCodIng = list(rqst)                             # el list() permite crear un arreglo list []
print(f'listaCodIng = {listaCodIng} \n')

rqst01 = rqst > 103                                  # Condicion, que datos voy a traer?, rspt 'cod Ing' > 105
print(f'Condicion codIng > 105: \n {rqst01} \n')

print(f'Datos Final:')
for i,u in enumerate(rqst01):                        # El enumerate sirve p crear un indice al arreglode datos.
    #print (f'indice = {u}, valor = {i}')
    # print(f'indice = {i}, truefalse = {u}, valor = {listaCodIng[i]}')
    if u == True:
        print(f'dato = {listaCodIng[i]}')
        
```
