```python
import pandas as pd
from matplotlib import pyplot as plt
puppies = pd.read_excel('D:/phytonSpace/DeepLearning/DataCamp/Hillary MML/officer_hours.xlsx', 'puppis')
plt.hist(puppies.weight)
plt.xlabel('Puppy Weight (lbs)')
plt.ylabel('Number of Puppies')
plt.show()
```

```python
# agrego bins= 50
import pandas as pd
from matplotlib import pyplot as plt
puppies = pd.read_excel('D:/phytonSpace/DeepLearning/DataCamp/Hillary MML/officer_hours.xlsx', 'puppis')
plt.hist(puppies.weight,bins=50)
plt.xlabel('Puppy Weight (lbs)')
plt.ylabel('Number of Puppies')
plt.show()
```

```python
# agrego un range de 5 a 35
import pandas as pd
from matplotlib import pyplot as plt
puppies = pd.read_excel('D:/phytonSpace/DeepLearning/DataCamp/Hillary MML/officer_hours.xlsx', 'puppis')
plt.hist(puppies.weight,range=(5,35),bins=50,)
plt.xlabel('Puppy Weight (lbs)')
plt.ylabel('Number of Puppies')
plt.show()
```

```python
#agrego Normalizacion Density=True
import pandas as pd
from matplotlib import pyplot as plt
puppies = pd.read_excel('D:/phytonSpace/DeepLearning/DataCamp/Hillary MML/officer_hours.xlsx', 'puppis')
plt.hist(puppies.weight,range=(5,35),bins=50,density=True)
plt.xlabel('Puppy Weight (lbs)')
plt.ylabel('Number of Puppies')
plt.show()
```
